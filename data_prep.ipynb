```python
import pandas as pd
```

```python
#Reading datasets
customers_df = pd.read_csv('olist_customers_dataset.csv')
order_items_df = pd.read_csv('olist_order_items_dataset.csv')
geolocation_df = pd.read_csv('olist_geolocation_dataset.csv')
order_payments_df = pd.read_csv('olist_order_payments_dataset.csv')
order_reviews_df = pd.read_csv('olist_order_reviews_dataset.csv')
orders_df = pd.read_csv('olist_orders_dataset.csv')
products_df = pd.read_csv('olist_products_dataset.csv')
sellers_df = pd.read_csv('olist_sellers_dataset.csv')

df_list = [customers_df, order_items_df, geolocation_df, order_payments_df,
           order_reviews_df, orders_df, products_df, sellers_df]
```

```python
#Checking for missing values
for df in df_list:
    missing_values = df.isnull().sum()
    print(missing_values[missing_values>0])
```

```python
#Handling missing values in 'order_reviews_df'
order_reviews_df['review_comment_title'] = order_reviews_df['review_comment_title'].fillna('Sem titulo')
order_reviews_df['review_comment_message'] = order_reviews_df['review_comment_message'].fillna('Sem mensagem')
order_reviews_df.head()
```

```python
#Handling missing values in 'orders_df'
missing_values = orders_df.isnull().sum()
missing_values = missing_values[missing_values>0]
mv_cols = missing_values.index
#orders_df[mv_cols].isnull().any()
for col in mv_cols:
    print(f"Linhas com valores faltantes na coluna '{col}':")
    missing_rows = orders_df[orders_df[col].isnull()]
    print(missing_rows)
    print("\n")
#missing_values = orders[orders.isna().any(axis=1)]
```
